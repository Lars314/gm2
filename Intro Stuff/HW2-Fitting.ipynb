```python
import ROOT as r
import numpy as np
```

```python
f = r.TFile("./gm2offline_ana_22530535_16119.00442.root")
f.ls()
```

```python
t = f.Get("farline/eventTree")
```

```python
def normalization(t, N_ini, muon_life, A, freq, phi):
      
    return N_ini * np.exp(-t/muon_life) * (1 - A * np.cos(freq*t + phi))
    
```

```python
norm = r.TF1("norm", "[0]*(TMath::Exp(-x/[1]))*(1-[2]*TMath::Cos([3]*x + [4]))", 30, 40)
```

```python
norm.SetParameter(0, 10 ** 5)
norm.SetParameter(1, 64)
norm.SetParameter(2, 1/3)
norm.SetParameter(3, 229073)
norm.SetParameter(4, 0)
```

```python
xbin = 4700
c = r.TCanvas("c","c", 1000, 500)
c.Divide(2,1)

c.cd(1)
h = r.TH2D("h_"+str(xbin), # name of the histogram in root
          "Energy vs. Time for "+str(xbin)+" time bins; Time [#mus]; Energy [MeV]", #histogram main title and axes titles, seperated by ; 
           xbin,0,700, # x-axis: number of bins, lower bound, upper bound
           100,0,6000 ) #same, but for y-axis
t.Draw("energy:time*1.25/1000.>>h_"+str(xbin),"","goff")
h.GetXaxis().SetRangeUser(30,40)
h.ProjectionX().Draw("hist") #projects this onto the x-axis
r.gPad.SetLogy() 

c.cd(2)
h = r.TH2D("h_"+str(xbin), # name of the histogram in root
          "Energy vs. Time for "+str(xbin)+" time bins; Time [#mus]; Energy [MeV]", #histogram main title and axes titles, seperated by ; 
           xbin,0,700, # x-axis: number of bins, lower bound, upper bound
           100,0,6000 ) #same, but for y-axis
t.Draw("energy:time*1.25/1000.>>h_"+str(xbin),"","goff")
h.GetXaxis().SetRangeUser(30,40)
h2 = h.ProjectionX().Clone("hist") #projects this onto the x-axis

h2.Fit("norm")
h2.Draw()
r.gPad.SetLogy()


c.Draw()
```
