```python
import ROOT as r
```

```python
f = r.TFile("./gm2offline_ana_22530535_16119.00442.root")
f.ls()
```

```python
f.cd("farline")
f.ls()
```

```python
allCaloEnergies = f.Get("farline/allCaloEnergies").Clone("allCaloEnergies")
allCaloEnergies.SetDirectory(0)

c = r.TCanvas()
allCaloEnergies.Draw()
c.SetLogy()
c.Draw()
```

```python
integral = allCaloEnergies.Integral()
print("Integral under the curve:", integral)

allCaloEnergies.Scale(1/integral)

newIntegral = allCaloEnergies.Integral()
print("New integral under the curve:", newIntegral)
```

```python
c = r.TCanvas()
allCaloEnergies.Draw()
c.SetLogy()
c.Draw()
```

```python
c = r.TCanvas()
allCaloEnergies.Draw("hist")  # play around with the parameters to make stuff look a certain way
allCaloEnergies.SetLineColor(7)  # whatever
c.SetLogy()
c.Draw()
```
