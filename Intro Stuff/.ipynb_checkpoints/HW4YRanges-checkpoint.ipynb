```python
import ROOT as r
import math
from BlindersPy3 import Blinders
from BlindersPy3 import FitType

getBlinded = Blinders(FitType.Omega_a, 'memes')
```

```python
chain = r.TChain("farline/eventTree");
chain.Add("./9day/gm2offline_ana_233128_16528.00035.root");
chain.Add("./9day/gm2offline_ana_233130_16535.00353.root");
chain.Add("./9day/gm2offline_ana_233134_16482.00307.root");
chain.Add("./9day/gm2offline_ana_233136_16528.00161.root");
```

```python
"""
Define the blinded wiggle function used for fitting
"""

def blinded_wiggle(x, p):
    norm = p[0]
    life = p[1]
    asym = p[2]
    R = p[3]
    phi = p[4]
    
    time = x[0]
    omega = getBlinded.paramToFreq(R)
    
    return norm * math.exp(-time/life) * (1 - asym*math.cos(omega*time + phi))

func = r.TF1("func", blinded_wiggle, 30, 600, 5)
func.SetParNames('N', '#tau', 'A', 'R( #omega_{a} )', '#phi')
func.SetParameters(9000, 64.4, 0.4, 0, 5.2)
func.SetLineColor(3)
func.SetNpx(10000)
```

```python
"""
Create a list of histograms of Y ranges for 
"""

rows = []
for i in range(0, 6):    
    hist = r.TH2D("y_"+str(i),
                  "Energy vs Time with Y ["+str(i)+","+str(i+1)+"]; Time [#mus]; Energy [MeV]",
                  4700, 0, 700, 100, 0, 6000)

    chain.Draw("energy:time*1.25/1000.>>y_"+str(i), "y>"+str(i)+" && y<"+str(i+1), "goff")
    rows.append(hist.Clone("y_"+str(i)))
```

```python
"""
Plot the histograms
"""
c = r.TCanvas("c","c", 1800, 1200)
c.Divide(3,2)

for i, hist in enumerate(rows):
    c.cd(i+1)
    hist.Draw("colz")
    r.gPad.SetLogz()

c.Draw()
```

```python
"""
Plot the fit 
"""

eLow = 1700
eHigh = 6000

c = r.TCanvas("c","c", 1800, 1800)
c.Divide(3,2)
r.gStyle.SetOptStat(0)

results = []

for i, hist in enumerate(rows):
    c.cd(i+1)
    
    binLow = hist.GetYaxis().FindBin(eLow)
    binHigh = hist.GetYaxis().FindBin(eHigh)
    
    hist.GetXaxis().SetRangeUser(30, 600)
    hist.ProjectionX("_px", binLow, binHigh).Draw("colz")
    results.append(hist.ProjectionX("_px", binLow, binHigh).Fit("func", "REMQS"))
    
    r.gPad.SetLogz()
c.Draw()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
"""
Put the results into a pandas dataframe to look at
"""

results_df = pd.DataFrame(columns = ['Y', 'N', '#tau', 'A', 'R #omega', '#phi'])

for index in range(0, 6):
    results_df.loc[index] = pd.Series({'Y':index,
                                       'N':results[index].Parameter(0),
                                       '#tau':results[index].Parameter(1),
                                       'A':results[index].Parameter(2),
                                       'R #omega':results[index].Parameter(3),
                                       '#phi':results[index].Parameter(4)})
```

```python
fig, ax = plt.subplots(1, 5)
fig.set_size_inches(25,5)

ax[0].plot(results_df['Y'], results_df['N'], linestyle='none', marker='o');
ax[0].set_xlabel("Row");
ax[0].set_ylabel("$N_0$");
ax[0].set_title("Spread of $N_0$ by Crystal Row");

ax[1].plot(results_df['Y'], results_df['#tau'], linestyle='none', marker='o');
ax[1].set_xlabel("Row");
ax[1].set_ylabel("$\\tau$");
ax[1].set_title("Spread of $\\tau$ by Crystal Row")

ax[2].plot(results_df['Y'], results_df['A'], linestyle='none', marker='o');
ax[2].set_xlabel("Row");
ax[2].set_ylabel("$A_0$");
ax[2].set_title("Spread of $A_0$ by Crystal Row")

ax[3].plot(results_df['Y'], results_df['R #omega'], linestyle='none', marker='o');
ax[3].set_xlabel("Row");
ax[3].set_ylabel("R($\omega_a$)");
ax[3].set_title("Spread of R($\omega_a$) by Crystal Row");

ax[4].plot(results_df['Y'], results_df['#phi'], linestyle='none', marker='o');
ax[4].set_xlabel("Row");
ax[4].set_ylabel("$\phi_0$");
ax[4].set_title("Spread of $\phi_0$ by Crystal Row")

fig.tight_layout();
```
