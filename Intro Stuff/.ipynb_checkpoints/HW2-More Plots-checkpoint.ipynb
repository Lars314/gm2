```python
import ROOT as r
f = r.TFile("./gm2offline_ana_22530535_16119.00442.root")
t = f.Get("farline/eventTree")
```

```python
c = r.TCanvas("c","c", 1000, 500)
c.Divide(2,1)

c.cd(1)
t.Draw("energy:x", "", "colz")
r.gPad.SetLogz()

c.cd(2)
t.Draw("energy:y", "", "colz")
r.gPad.SetLogz()

c.Draw()
```

```python
c = r.TCanvas("c","c", 1000, 500)
c.Divide(2,1)

c.cd(1)
t.Draw("x:time", "", "colz")
r.gPad.SetLogz()

c.cd(2)
t.Draw("x:time", "", "colz")
r.gPad.SetLogz()
r.gPad.SetLogx()

c.Draw()
```

```python
c = r.TCanvas("c", "c", 1800, 4800)
c.Divide(3,8)
c.Draw()

for i in range(1, 25):
    c.cd(i)
    caloHits = f.Get("farline/calo"+str(i)+"/hitPositions").Clone("hitPositions")
    caloHits.Draw()
    c.Update()

```
