```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dfs = []
ens = []
for epoch_num in range(50, 250, 50):
    dfs.append(pd.read_csv('./e' + str(epoch_num) + '.csv'))
    ens.append(epoch_num)
    
dfs.append(pd.read_csv('./e20.csv'))
ens.append(20)
```

```python
fig, ax = plt.subplots(1, 2)
fig.set_size_inches(14, 7)

for i in range(0, len(ens)):
    ax[0].plot(dfs[i]['accuracy'], label=str(ens[i]) + ' epochs')
    ax[1].plot(dfs[i]['loss'], label=str(ens[i]) + ' epochs')

    
ax[0].set_ylabel('accuracy %', size=22)
ax[1].set_ylabel('loss', size=22)
ax[0].set_xlabel('epoch', size=22)
ax[1].set_xlabel('epoch', size=22)
ax[0].legend(fontsize='large')
ax[1].legend(fontsize='large')

plt.savefig('./epoch_scan.png')
```
