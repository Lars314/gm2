```python
fig, ax = plt.subplots(1,3)
fig.set_size_inches(30,10)

df = pd.read_csv('metrics/confusion_knn.csv', sep=',', index_col=0)
grid = np.array(df.values)

im = ax[0].imshow(grid)

for k in range (0, len(grid)):
        for l in range (0, len(grid[0])):
            ax[0].text(l-0.35, k, str(grid[k][l])[:4], size=20, color='white')
            
            

df = pd.read_csv('metrics/confusion_rfc.csv', sep=',', index_col=0)
grid = np.array(df.values)

im = ax[1].imshow(grid)

for k in range (0, len(grid)):
        for l in range (0, len(grid[0])):
            ax[1].text(l-0.35, k, str(grid[k][l])[:4], size=20, color='white')
            
            
df = pd.read_csv('metrics/confusion_mlp.csv', sep=',', index_col=0)
grid = np.array(df.values)

im = ax[2].imshow(grid)

for k in range (0, len(grid)):
        for l in range (0, len(grid[0])):
            ax[2].text(l-0.35, k, str(grid[k][l])[:4], size=20, color='white')
```
